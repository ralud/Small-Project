## NYC Restaurant Health Inspection Grades and Cuisine Characteristics

For this project, I explored the relationship between [NYC restaurant inspection grades](http://www.nyc.gov/html/doh/html/services/restaurant-inspection.shtml) (the A, B,C, Grade Pending) and characteristics of the cuisines to which the restaurant belong to.

The hypotheses that I wanted to test is that foreign food safety and hygene norms affect restaurant health grades in NYC.  

In particular, I explore whether NYC health inspection scores (avialble from NYC Open Data) are correlated with the GDP per capita of the countries from which the cuisines originate. In the absence of cross-country data on food hygene standards, GDP per capita serves as a rough proxy.

Data Sources
* [NYC Restaurant Inspection Results](https://nycopendata.socrata.com/Health/Restaurant-Inspection-Results/4vkw-7nck); 
* [Countries GDP per Capita](http://data.worldbank.org/data-catalog/world-development-indicators)

Variables of Interest
* Restaurant Score: sums the number and severity of violations on a given inspection (Grade A: 0-13; Grade B: 14-27;Grade C: 28+)
* GDP per capita

```python
#import NYC restaurant inspection data downloaded from NYC Open Data
inspections=pd.read_csv("/Users/Raluca/Documents/project/WebExtract.txt", delimiter=",", header=0)
#import the cuisine description data downloaded from NYC Open Data
cuisine=pd.read_csv("/Users/Raluca/Documents/project/Cuisine.txt", delimiter=",", header=0)
#import the GDP per capita data downloaded from World Bank Development Indicators
gdpcapita=pd.read_csv("/Users/Raluca/Documents/project/gdp_cap_short.csv", delimiter=",", header=0)
#import the correspondence between cuisine descriptions and country names (i.e Chinese=China)
coresp=pd.read_csv("/Users/Raluca/Documents/project/code_ctry_coresp.csv", delimiter=",", header=0)
```

#Cleaning Data

```python
#transform the column names from uppercase to lowercase
inspections.columns = map(str.lower, inspections.columns)
#transform the grade date column to string type for further manipulation
inspections['gradedate'] = inspections['gradedate'].astype(str)
```

```python
#clean the grade dates
clean_gradedate = [pd.to_datetime(r.split(' ')[0], format='%Y-%m-%d') for r in inspections.gradedate]
inspections['gradedate'] = clean_gradedate
```

```python
#extract a smaller dataset containing the variables of interest 
#camis=restaurant code; cuisinecode=unique code assigned to a cuisine; gradedate=date when a restaurant grade was assigned)
#score=restaurant inspection score based on which grades (A, B, C and Grade Pending) are assigned
#insp_short=inspections[['camis','cuisinecode','gradedate','score','currentgrade']]
insp_short=inspections
#drop duplicate rows (many duplicates because unit of observation is restaurant-inspection-violation code)
insp_short.drop_duplicates()
#other basic cleaning
insp_short['currentgrade'].replace('P', 'Grade Pending', inplace = True)
insp_short['currentgrade'].replace('Z', 'Grade Pending', inplace = True)
insp_short['gradeyear']=pd.DatetimeIndex(pd.to_datetime(insp_short.gradedate, infer_datetime_format=True)).year
insp_short['grademonth']=pd.DatetimeIndex(pd.to_datetime(insp_short.gradedate, infer_datetime_format=True)).month
```

```python
#transform the column names from uppercase to lowercase
cuisine.columns = map(str.lower, cuisine.columns)
```

```python
#merge all the datasets together 
data=pd.merge(insp_short,cuisine, on='cuisinecode',how='inner')
data=pd.merge(data,coresp, on='codedesc',how='inner')
data=pd.merge(data,gdpcapita, on='country',how='inner')
```

The Relationship Between Restaurant Scores and Level of Development in Origin Country

```python
#Scatter plot of GDP per capita and Restaurant Scores
plt.figure(figsize=(10,5))
plt.scatter(data.gdp_2012.apply(np.log), data.score.apply(np.log),alpha=.04, lw=0)
plt.xlabel("Country (cuisine of origin) Income per Capita")
plt.ylabel("Restaurant Health Inspection Score")
plt.title("Restaurant Health Inspection Score and Country of Origin GDP per Capita",fontsize='x-large')
```

```python
#drop all the observations with missing score
data['score'].dropna(inplace=True)
#because the score can be zero and in the end we will want to log it we add 1 to the entire series
data['score']=data['score']+1
```

```python
#Because when using all the data the relationship is hard to see, we can average the restaurant scores at the cuisine level
cusine_counts = data['country'].value_counts()
cusine_counts
cuisine_score = data.groupby('cuisinecode').mean()
```

```python
# This relationship looks weakly downward-sloping
plt.figure(figsize=(7,7))
plt.scatter(cuisine_score.gdp_2013.apply(np.log), cuisine_score.score, color='red')
plt.xlabel("Country (cuisine of origin) Income per Capita")
plt.ylabel("Restaurant Health Inspection Score")
plt.title("Restaurant Health Inspection Score and Country of Origin GDP per Capita",fontsize='x-large')
```

```python
#To investigate whether the relationship between score and income per capita is statistically significant, we run a simple OLS regression
import statsmodels.formula.api as smf
data['score'].dropna(inplace=True)
data['gdp_2012'].dropna(inplace=True)
data['log_score'] = np.log(data['score'])
data['log_gdp'] = np.log(data['gdp_2012'])
```

```python
log_model = smf.ols(formula='log_score ~ log_gdp', data=data)
log_results = log_model.fit()
print 'LOG-LOG FIT SUMMARY'
print(log_results.summary())
```

The coefficient on log GDP per capita is negative and statistically significant suggesting that cuisines originating in countries with higher GDP per capita obtain lower scores (i.e have fewer health violations on restaurant inspections).

Conclusion: Restaurants belonging to cuisines which originate from more developed countries (as proxied by GDP per capita) score better on NYC restaurant health inspections. This provides suggestive evidence that food hygene norms persist in a different institutional environment (United States).
